```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("images/image2.png")
```

```python
cv2.imshow("image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img)
```

```python
img1 = cv2.line(img,(100,200),(400,500),(255,255,0),5)
plt.imshow(img1)
```

```python
img1 = cv2.rectangle(img,(100,200),(400,500),(255,255,0),5)
plt.imshow(img1)
```

```python
img1 = cv2.circle(img,(100,200),50,(255,255,0),5)
plt.imshow(img1)
```

```python
img1 = cv2.putText(img,"Hello World!!!", (150,200), cv2.FONT_HERSHEY_PLAIN, 3, (0,0,0),5)
plt.imshow(img1)
```
