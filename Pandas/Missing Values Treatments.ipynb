```python
print(df["Reviews"].value_counts())
```

```python
data = df["Installs"]

lst = []
num = 0;

for i in data:
    filter_val = float(i.replace(',','').replace('+',''))
    df.loc[i] = filter_val
    num = num + 1
    lst.append(filter_val) 
```

```python
## Apply Function

def filter_data(x):
    x = float(x.replace(',','').replace('+',''))
    return x

df["Installs"].apply(filter_data)
```
