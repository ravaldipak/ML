```python
def is_odd(val):
  return val % 2 != 0

def divide_compare(str_val):
  middle_idx = len(str_val) // 2
  first_str, second_str = str_val[:middle_idx], str_val[middle_idx:][::-1]
  for a, b in zip(first_str, second_str):
    if a != b:
      return False
  return True


def is_palindrom(str_val):
  if not is_odd(len(str_val)):
    return divide_compare(str_val)
  else:
    mid_idx = len(str_val) // 2
    new_str = str_val[:mid_idx] + str_val[mid_idx+1:]
    return divide_compare(str_val)
```
