```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as hsc
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
df
```

```python
X = df.iloc[:,[3,4]].values
```

# Single Linkage

```python
Z = hsc.linkage(X,method="single")
```

```python
plt.figure(figsize=(10,10))
dendogram = hsc.dendrogram(Z)
plt.show()
```

# Complate Linkage

```python
Z = hsc.linkage(X,method="complete")
plt.figure(figsize=(10,10))
dendogram = hsc.dendrogram(Z)
line = plt.axhline(y=100,lw=3)
plt.show()
```

# Avarage Linkage

```python
Z = hsc.linkage(X,method="average")
plt.figure(figsize=(10,10))
dendogram = hsc.dendrogram(Z)
line = plt.axhline(y=100,lw=3)
plt.show()
```

# Agglorimatrive clustering

```python
hc = AgglomerativeClustering(n_clusters = 3 , linkage="complete")
```

```python
y_hc = hc.fit_predict(X)
```

```python
y_hc
```

## Visulization

```python
plt.scatter(X[y_hc==0, 0], X[y_hc==0, 1], s=100, c='red', label ='Cluster 1')
plt.scatter(X[y_hc==1, 0], X[y_hc==1, 1], s=100, c='blue', label ='Cluster 2')
plt.scatter(X[y_hc==2, 0], X[y_hc==2, 1], s=100, c='green', label ='Cluster 3')
plt.scatter(X[y_hc==3, 0], X[y_hc==3, 1], s=100, c='purple', label ='Cluster 4')
plt.scatter(X[y_hc==4, 0], X[y_hc==4, 1], s=100, c='orange', label ='Cluster 5')
plt.title('Clusters of Customers (Hierarchical Clustering Model)')
plt.xlabel('Annual Income(k$)')
plt.ylabel('Spending Score(1-100)')
plt.show()
```
