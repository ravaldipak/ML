```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Importing time series data

```python
google = pd.read_csv("google_stock_market.csv", index_col="Date", parse_dates=["Date"])
google.head()
```

```python
humidity = pd.read_csv("humidity.csv", index_col="datetime", parse_dates=['datetime']);
humidity.head()
```

### Cleaning and preparing time series data¶

```python
google.isnull().sum()
```

Google stocks data doesn't have any missing values but humidity data does have its fair share of missing values. It is cleaned using fillna() method with ffill parameter which propagates last valid observation to fill gaps

```python
humidity = humidity.iloc[1:]
humidity = humidity.fillna(method="ffill")
humidity.head()
```

```python
humidity.isnull().sum()
```

### Visualizing the datasets

```python
humidity["Kansas City"].asfreq("M").plot()
plt.title("Humidity in Kansas City")
plt.show()
```

```python
google["2008":"2010"].plot(subplots=True, figsize=(10,12))
plt.title("Google Stock Attribute from 2008 to 2010")
plt.show()
```
